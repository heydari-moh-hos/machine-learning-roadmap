## 3. Conditional selection

```python
# retail.Country.value_counts()
retail['Country'].value_counts()
```

```python
retail.Country == 'USA'
```

```python
retail[retail.Country == 'USA']
```

```python
retail[retail.Country == 'USA'].reset_index(drop = True)
```

```python
# retail[retail.Country == 'USA']
retail.loc[retail.Country == 'USA']
```

```python
retail.loc[retail.Country == 'USA',['Quantity','Price']]
```

```python
# (retail.Country == 'USA') & (retail.Price >= 2)
# if 'and' is applied instead of '&' wiil raise error
# if '()' is not applied will raise error

retail.loc[(retail.Country == 'USA') & (retail.Price >= 2)].reset_index(drop=True)
```

```python
retail.loc[(retail.Country == 'USA') & (retail.Price >= retail.Price.mean())].reset_index()
```

```python
retail.loc[(retail.Country == 'USA') | 
           (retail.Price >= retail.Price.mean())].reset_index()
```

```python
# retail.Country == 'USA' | retail.Country == 'Germany' 
# retail.Country.isin(['USA','Germany'])
retail.loc[retail.Country.isin(['USA','Germany']),['Invoice','Quantity','Country']]
```

```python
retail.loc[retail.Quantity < 0,]
```

```python
retail.loc[retail.Price < 0]
```

```python
import numpy as np
# in excel: IF(codition, value if condition True, value if condition False)

retail['new_col'] = np.where(retail["Price"] < 2, "low", "high")
retail
```
